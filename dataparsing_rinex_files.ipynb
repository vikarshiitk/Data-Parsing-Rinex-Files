# Estimating Trajectory of Car

```python


import numpy as np
import sympy as sym
sym.init_printing()
from sympy import *


# In[2]:


X0=np.matrix([-9.75,-9.75,0,0],dtype="float")
X0=X0.T
X0


# In[3]:


x, y ,Vx, Vy, t=sym.symbols("x y Vx Vy t")


# In[4]:


x_=x+Vx*t
y_=y+Vy*t
Vx_=Vx
Vy_=Vy


# In[5]:


F=np.matrix([[sym.diff(x_,x),0,sym.diff(x_,Vx).subs([(t,1)]),0],
             [0,sym.diff(y_,y),0,sym.diff(y_,Vy).subs([(t,1)])],
             [0,0,sym.diff(Vx_,Vx),0],
             [0,0,0,sym.diff(Vy_,Vy)]],dtype="float")
F


# In[6]:


X_N_m=F*X0
X_N_m


# In[7]:


P0=np.eye(4)
P0=np.matrix(P0)
P0


# In[8]:


P_N_m=F*P0*(F.T)
P_N_m


# In[9]:


Y_=np.matrix([[9.72,9.72,22.08,22.08],
       [8.77,8.77,20.74,20.74],
       [7.97,7.97,19.42,19.42],
       [7.34,7.34,18.1,18.1],
       [6.93,6.93,16.81,16.81],
       [6.79,6.79,15.53,15.53],
       [6.93,6.93,14.28,14.28],
       [7.34,7.34,13.06,13.06],
       [7.97,7.97,11.88,11.88],
       [8.77,8.77,10.76,10.76],
       [9.72,9.72,9.72,9.72],
       [10.76,10.76,8.77,8.77],
       [11.88,11.88,7.97,7.97],
       [13.06,13.06,7.34,7.34],
       [14.28,14.28,6.93,6.93],
       [15.53,15.53,6.79,6.79],
       [16.81,16.81,6.93,6.93],
       [18.1,18.1,7.34,7.34],
       [19.42,19.42,7.97,7.97],
       [20.74,20.74,8.77,8.77],
       [22.08,22.08,9.72,9.72]])
Y_


# In[10]:


C=np.matrix([[-10,0],
             [0,-10],
             [10,0],
             [0,10]])
C


# In[11]:


Rk=np.identity(4)
Rk


# In[12]:


f1=((C[0,0]-x)**2+(C[0,1]-y)**2)**(0.5)
f2=((C[1,0]-x)**2+(C[1,1]-y)**2)**(0.5)
f3=((C[2,0]-x)**2+(C[2,1]-y)**2)**(0.5)
f4=((C[3,0]-x)**2+(C[3,1]-y)**2)**(0.5)


# In[13]:


Cx,Cy,X,Y,r=sym.symbols("Cx Cy X Y r")


# In[14]:


fm=((Cx-X)**2+(Cy-Y)**2)**(0.5)
fm


# In[15]:


I=np.identity(4)
I


# In[16]:


X_final=[]
Y_final=[]
trace=[]
for j in range(21):  
    X_N_m=F*X0
    P_N_m=F*P0*(F.T)
    Y_m=np.zeros((len(C),1))
    for i in range(len(C)):
        Y_m[i][0]=fm.subs([(Cx,C[i,0]),(Cy,C[i,1]),(X,X0[0,0]),(Y,X0[1,0])])
    Y_m=np.matrix(Y_m)
    H=np.matrix([[sym.diff(f1,x).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),sym.diff(f1,y).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),0,0],
            [sym.diff(f2,x).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),sym.diff(f2,y).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),0,0],
            [sym.diff(f3,x).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),sym.diff(f3,y).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),0,0],
            [sym.diff(f4,x).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),sym.diff(f4,y).subs([(x,X_N_m[0,0]),(y,X_N_m[1,0])]),0,0]],dtype="float")
    Kk=P_N_m*(H.T)*(H*P_N_m*(H.T)+Rk)**(-1)
    Y_g=np.matrix(Y_[j])
    Y_g=Y_g.T
    X_N_p=X_N_m+Kk*(Y_g-Y_m)
    P_N_p=(I-(Kk*H))*P_N_m
    
    X0=X_N_p
    P0=P_N_p
    
    X_final.append(X0[0,0])
    Y_final.append(X0[1,0])
    trace.append(float(P0.trace()))


# In[17]:


import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,10))
plt.plot(X_final,Y_final,color="yellow")
plt.scatter(X_final,Y_final,color="blue")


# In[18]:


T=np.arange(21)
fig,ax=plt.subplots(figsize=(10,10))
plt.plot(T,trace,color="yellow")
plt.scatter(T,trace,color="blue")

```
